#### Final sales data

```python
# Now we can run sales.csv file to get all months of data

df_all = pd.read_csv('sales.csv')

df_all.head()
```

```python
df_all.shape
```

### Clean up the data

```python
# checking data types of each column
df_all.info()
```

```python
# Checking null values
df_all.isnull().sum()
```

```python
# Drop all the null values
df_all.dropna(inplace=True)
```

```python
# overview of the data
df_all.describe()
```

```python
# Changing column names

new_column = {'Order ID':'Order_ID','Quantity Ordered':'Quantity_Ordered','Price Each':'Price_Each','Order Date':'Order_Date','Purchase Address':'Address'}

df_all.rename(columns=new_column,inplace=True)
```

```python
# dropping duplicate values

df_all.drop_duplicates(inplace=True)
```

```python
# changing 'Order Date' type to datatime so we can do more analysis

df_all['Order_Date'] = pd.to_datetime(df_all['Order_Date'],errors='coerce')
```

```python
# creating new month column
df_all['order_month'] = df_all['Order_Date'].dt.month_name().str[0:3]
```

```python
str_rows = df_all[df_all['Order_ID'].str[0:2] == 'Or']
str_rows.head()
```

```python
# We need to remove that row 
df_all = df_all[df_all['Order_ID'].str[0:2] != 'Or']
```

```python
# changing price column to numeric so we can calculate total sales

df_all['Price_Each'] = pd.to_numeric(df_all['Price_Each'])

df_all['Quantity_Ordered'] = pd.to_numeric(df_all['Quantity_Ordered'])
```

```python
# creating new sales column so we can find best sales month 

df_all['sales'] = df_all['Quantity_Ordered'] * df_all['Price_Each']
```

```python
df_all.head()
```

### 1. Finding which month had the highest number of sales

```python
# finding what was the best month for sales and how much earned that month

month_sales = df_all[['order_month','sales']].groupby('order_month').sum().sort_values('sales', ascending=False)

month_sales
```

```python
month_sales.reset_index(inplace=True)
month_sales.head()
```

```python
# Top sales per month using plots 
plt.figure(figsize=(8,4))

sns.barplot(data=month_sales, x='order_month', y='sales')

plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Total Sales per Month")

plt.show()
```

* `December` month sales is higher than any other month sales.

### 2. Which city had the highest number of sales 

```python
# Creating new city column
#df_all['city'] = df_all['Address'].str.split(",")[1]
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

df_all['city'] = df_all['Address'].apply(lambda x: f'{get_city(x)} ({get_state(x)})')
```

```python
df_all.head()
```

```python
# finding which city had highest sales

city_sales = df_all[['city','sales']].groupby('city').sum()

city_sales
```

```python
city_sales.reset_index(inplace=True)
```

```python
# Top sales per month using plots 
plt.figure(figsize=(17,7))

sns.barplot(data=city_sales, x='city', y='sales')

plt.xlabel("Cities",fontsize=14)
plt.ylabel("Total Sales",fontsize=14)
plt.title("Total Sales per city($)",fontsize=14)

plt.show()
```

* As the results `San Francisco` and `Los Angeles` and `New york` are top 3 cities for total sales.

### 3. What time should we display advertisements to maximize likelihood of customer's buying

### 4. What products are most often sold together

```python
df_all.head()
```

```python
# creating new column that joins all the products that have same order ID

df = df_all[df_all['Order_ID'].duplicated(keep=False)].copy()

df['products'] = df.groupby('Order_ID')['Product'].transform(lambda x:', '.join(x))

df.head()
```

```python
# Order-id have duplicate entries so getting unique ID

df = df[['Order_ID','products']].drop_duplicates()
df.head()
```

```python
# importing requried libraires
from itertools import combinations
from collections import Counter
```

```python
# for reference how this libray will work

count = Counter()

items = [[4,5],[4,5,6],[5,6,7]]

for item in items:
    count.update(Counter(combinations(item, 2)))
    
print(count)    
    
```

```python
for row in df['products']:
    print(row)
```

```python
# Most commonly sold 2 things together

for row in df['products']:
    # this will make a list ['Google Phone', ' Wired Headphones'] like this
    items = row.split(',')
    count.update(Counter(combinations(items, 2))) # 2 means how things we want to find 
    
    
for key,value in count.most_common(10):
    print(key,value)
    
```

* 'iPhone', ' Lightning Charging Cable' are sold together 1002 times.
* Most of them are phones, so If customer bought phone there is higher chance to buy some other things as well. 

### 5. what product sold the most?

```python
df_all.head()
```

```python
# We can get this by grouping the products and sum value of the quantity

product_ordered = df_all.groupby('Product')['Quantity_Ordered'].sum().sort_values(ascending=False)
product_ordered
```

```python
product_count = df_all.groupby('Product').agg(
    quantity = pd.NamedAgg(column='Quantity_Ordered',aggfunc=sum)).reset_index()

product_count.head()
```

```python
# Top 10 sold products

plt.figure(figsize=(25,8))

sns.barplot(data=product_count.head(10), x='Product', y='quantity')

plt.xlabel("Products",fontsize=14)
plt.ylabel("Total counts",fontsize=14)
plt.xticks(rotation= 'vertical', size=15)
plt.title("Number of times a product has been sold",fontsize=14)

plt.show()
```

* `AAA Batteries(4-pack)` sold most of the times.

```python
prices = df_all.groupby('Product').agg(Avg_price=('Price_Each', np.mean)).reset_index()

prices.head()
```

```python
# For to make a visualizations 
product_df = product_count.head(10)
avg_price = prices.head(10)
```

```python
# Total quantity with avg price of each product

fig, ax1 = plt.subplots(figsize=(25,10))

ax2 = ax1.twinx()

# Plot barplot
ax1.bar(product_df['Product'],product_df['quantity'], color='blue', alpha=0.5, label='Barplot')

# Plot lineplot
ax2.plot(avg_price['Product'], avg_price['Avg_price'], color='red', marker='o', linestyle='-', label='Lineplot')

# Set labels and title
ax1.set_xlabel('Product name',fontsize=20)
ax1.set_ylabel('Quantity ordered',color='blue',fontsize=20)
ax2.set_ylabel('Price $',color='red',fontsize=20)
ax2.set_xticks(range(len(product_df['Product'])))
ax2.set_xticklabels(product_df['Product'],rotation='vertical',size=8)
ax1.set_title('Total quantity with avg price of each product',fontsize=20)

# Show the plot
plt.show()
```
